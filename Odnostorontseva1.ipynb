```python
import pandas as pd #Импорт библиотеки для работы с таблицами
data=pd.read_csv('/Data.csv') #загрузка датасета
data #Отображение датасета
```

```python
data['iPhone'] = data['iPhone'].str[1:] #Удаление первого символа у колонки айфонов, чтобы не было символа доллара
data.drop(labels = [71],axis = 0,inplace=True) #Удаление строки 71, тк она чем-то не понравилась колабу
data['iPhone'] = data['iPhone'].astype (float) #Изменяем тип данных по всему столбцу на флот
```

```python
# Переименование стран которые не "Приклеились"
data.at[7, 'Country'] = 'United States of America'
data.at[10, 'Country'] = 'United Kingdom'
data.at[48, 'Country'] = 'Czechia'
data.at[26, 'Country'] = 'Dominican Rep.'
data.at[56, 'Country'] = 'Uruguay'
```

```python
data.head() #Просмотр датасета
```

```python
#"Левое" склеивание данных мировой карты и данных датасета, отображение тех, что не склеились
new = data.merge(world, how='left', left_on=['Country'], right_on=['name'])
new.loc[new['continent'].isnull()] 
```

```python
#Удаляем оставшиеся страны, так как их почему-то нет в геопандасе
data.drop(labels = [11,66],axis = 0,inplace=True)
```

```python
#Повторная проверка на "Несклеившиеся" страны, датасет готов
new = data.merge(world, how='left', left_on=['Country'], right_on=['name'])
new.loc[new['continent'].isnull()]
```

```python
#Закидываем в переменную данные для склеивания, Страна и айфон (по первому датасету)
done = world.merge(new[['iPhone','Country']], how = 'left', left_on='name', right_on= 'Country')
```
